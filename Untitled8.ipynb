```python





#Import Libraries
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from google.colab import drive
drive.mount('/content/drive')
base_dir = '/content/drive/MyDrive/Dataset'


#  Load Images

IMG_SIZE = 64
data = []

def load_images_from_folder(folder, label):
    for file in os.listdir(folder):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            path = os.path.join(folder, file)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            data.append((img, label))

# Subfolders 0–9 (labeled)
for label in range(10):
    folder_path = os.path.join(base_dir, str(label))
    if os.path.exists(folder_path):
        load_images_from_folder(folder_path, label)

# Main folder (unlabeled)
for file in os.listdir(base_dir):
    if file.lower().endswith(('.png', '.jpg', '.jpeg')):
        path = os.path.join(base_dir, file)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        data.append((img, -1))

print(f" Total images found (labeled + unlabeled): {len(data)}")


#  Prepare Data

labeled_data = [d for d in data if d[1] != -1]
X = np.array([i[0] for i in labeled_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 1) / 255.0
y = np.array([i[1] for i in labeled_data])
y_cat = to_categorical(y, num_classes=10)


# Train-Test Split (70:30)

X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.3, random_state=42, stratify=y)


# CNN Model

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()


#Train Model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=32)


# Evaluate

y_pred = np.argmax(model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)

acc = accuracy_score(y_true, y_pred)
print(f" Overall Accuracy: {acc*100:.2f}%")


#  Confusion Matrix

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()


#  Performance Table

report = classification_report(y_true, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
report_df['accuracy'] = acc  # overall accuracy

# Combine confusion matrix and performance into one Excel
excel_path = '/content/drive/MyDrive/Performance.xlsx'
with pd.ExcelWriter(excel_path) as writer:
    pd.DataFrame(cm).to_excel(writer, sheet_name='Confusion_Matrix', index=True)
    report_df.to_excel(writer, sheet_name='Performance_Table')

print(f" Confusion Matrix & Performance Table saved at: {excel_path}")



print(" generated successfully!")




```
